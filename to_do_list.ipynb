```python
# Simple CLI To-Do List App

tasks = []     #empty list

def add_task():  # adding new tasks to the list
    task = input("Enter a new task: ")
    tasks.append(task)
    print("Task added successfully!\n")

def view_tasks():    #display added tasks
    if len(tasks) == 0:
        print("No tasks available.\n")
    else:
        print("\nYour To-Do List:")      #list start from 1
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")
        print()

def remove_task():    #remove task from the list
    view_tasks()

    if len(tasks) == 0:
        return

    task_number = int(input("Enter the task number to remove: "))
    tasks.pop(task_number - 1)
    print("Task removed successfully!\n")

while True:
    print("==== TO-DO LIST MENU ====")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        add_task()

    elif choice == "2":
        view_tasks()

    elif choice == "3":
        remove_task()

    elif choice == "4":   #exit the list
        print("Goodbye!")
        break

    else:
        print("Invalid choice\n")
```
